```python
#1:モジュールのインポート
import glob
import matplotlib.pyplot as plt
import cv2
from mpl_toolkits.axes_grid1 import ImageGrid
%matplotlib inline
```

輝度

```python
file = glob.glob("*.jpg")

for x in  file:
  file_name = x.split("/")[-1]
  img=cv2.imread(file_name)
  
  hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS) # hls票色系に変換
  h,l,s = cv2.split(hls) # 各成分に分割
  
  # plt.figure()　前に置くとle6だけなぜか出る
  plt.hist(l.ravel(), 256,[0,256], alpha=0.3, histtype='stepfilled', color='r')
  # plt.figure()

  
plt.show()
```
